```python
import re

COLOR_GREEN = "\033[32m"
COLOR_RED   = "\033[31m"
RESET       = "\033[0m"

strings = ["wazzzzzup", "wazzzup", "wazup"]
pattern = r"..zzz.."

def highlight_match(string, pattern, color):
    result = re.search(pattern, string)
    start  = result.start()
    end    = result.end()

    before  = string[:start]
    match   = string[start:end]
    after   = string[end:]

    highlighted = before + color + match + RESET + after
    print(string + ": " + highlighted)

matches  = []
skipped  = []

for string in strings:
    if re.search(pattern, string):
        matches.append(string)
    else:
        skipped.append(string)

print("Match:")
for string in matches:
    highlight_match(string, pattern, COLOR_GREEN)

print("\nSkip:")
for string in skipped:
    print(string + ": " + string)
```
