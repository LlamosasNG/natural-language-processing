```python
import re

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "The cat and the dog are playing in the garden",
    "The dog is chasing the ball in the park",
    "I love playing football and basketball",
    "Football is the best sport in the world",
    "Python is a great programming language",
    "I love programming in Python and Java"
]

print(f"Number of documents: {len(documents)}")
for i, doc in enumerate(documents):
    print(f"  D{i+1}: {doc}")  


def preprocess(text):
    text = text.lower()                      
    text = re.sub(r'[^\w\s]', '', text)          
    tokens = text.split()                        
    return ' '.join(tokens)


documents_clean = [preprocess(doc) for doc in documents]
print(f"Preprocessed example:\n{documents_clean[0]}")

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents_clean)

tfidf_vocabulary = vectorizer.get_feature_names_out()
print(f"Vocabulary size: {len(tfidf_vocabulary)}")
print(f"Vocabulary: {list(tfidf_vocabulary)}")

tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray().round(3),
    columns=tfidf_vocabulary,
    index=[f"D{i+1}" for i in range(len(documents))]
)
print("\nTF-IDF Matrix:\n")
print(tfidf_df.to_string())
```
