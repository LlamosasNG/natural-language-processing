```python
import re

words = ["Python", "Alice", "hello", "SQL", "John", "Europe", "Bob"]
for w in words:
    match = re.fullmatch(r'[A-Z][a-zA-Z]*(?<![aeiouAEIOU])', w)
    print(f"  {w:>10} → {'✅' if match else '❌'}")
```

```python
for w in words:
    match2 = re.fullmatch(r'[A-Z][a-zA-Z]*[^aeiouAEIOU\W]\b', w)
    print(f"  {w:>10} → {'✅' if match2 else '❌'}")
```
