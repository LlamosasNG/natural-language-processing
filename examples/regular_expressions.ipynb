```python
import re

# Rule 2: Common Regex Tokens in Python
re.findall(r"\w+", "Hola Mundo 123")
```

```python
# Rule 3: Searching vs Matching
print(re.search(r"\d+", "ID: 42"))

print(re.match(r"\d+", "42 is the answer")) # ✅
print(re.match(r"\d+", "ID: 42 ")) # ❌
```

```python
# Rule 4: Full Validation Requires Full Match
print(re.fullmatch(r"-?\d+", "123"))
print(re.fullmatch(r"-?\d+", "-45"))

pattern = r"-?\d+"

integer = re.compile(pattern)
print(bool(integer.fullmatch("123")))    # True
print(bool(integer.fullmatch("-45")))    # True
print(bool(integer.fullmatch("abc123"))) # False
print(bool(integer.fullmatch("12a")))    # False
print(bool(integer.fullmatch("  7  ")))  # False
print(bool(integer.fullmatch("7")))      # True

integer = re.compile(r"^-?\d+$")
print(bool(integer.match("123")))    # True
print(bool(integer.match("-45")))    # True
print(bool(integer.match("12a")))    # False
print(bool(integer.match("x12")))    # False
print(bool(integer.match("12 ")))    # False
print(bool(integer.match("-45")))    # True
```

```python
# Rule 5: Capturing Groups
# Numbered Groups
match = re.search(r"(\w+)@(\w+)\.(\w+)", "test@email.com")

print(match.group(1)) # test
print(match.group(2)) # email
print(match.group(3)) # com

# Named Groups (Recommended)
pattern = r"(?P<user>\w+)@(?P<domain>\w+)\.(?P<tld>\w+)"
match = re.search(pattern, "test@email.com")

print(match.group("user"))
print(match.group("domain"))
```

```python
# Rule 6: findall() vs finditer()
# findall()
print(re.findall(r"\d+", "A1 B22 C333")) # ['1','22','333']

# finditer()
for m in re.finditer(r"\d+", "A1 B22"):
	print(m.group(), m.start())
```

```python
# Rule 7: Substitution with re.sub()
# Replace sensitive data
print(re.sub(r"\d{4}-\d{4}", "XXXX-XXXX", "Card: 1234-5678"))

# Normalize whitespace
print(re.sub(r"\s+", " ", "Too     many           spaces"))
```

```python
# Rule 8: Greedy vs Non‑Greedy
# Greedy (default)
print(re.search(r"<.*>", "<tag>content</tag>"))

# Non‑greedy
print(re.search(r"<.*?>", "<tag>content</tag>"))
```

```python
#Rule 9: Flags Modify Behavior
# Example: case‑insensitive
print(re.search(r"python","Python", re.IGNORECASE))

text = "Hello World\nThis is a example regex"
# Multiline mode:
print(re.search(r"^Hello", text, re.MULTILINE))
```
